```python
import matplotlib.pyplot as plt             # to visualize
from tabulate import tabulate               # to print the table
import matplotlib as mat                    # to visualize 
import seaborn as sns                       # to visualize
import pandas as pd                         # for data reading
import numpy as np        
```

```python
# Load the merged data from an Excel file
file_df = pd.read_csv(r'C:\Users\Lenovo/football_merged_file.csv')
file_df
```

```python
PERFORMANCE ANALYSIS
```

```python
Top 5 players based on red card count
```

```python
# Group by player_name and sum the red cards
top_players = file_df.groupby('name')['red_cards'].sum().reset_index()

# Sort the players by total red cards in descending order
top_players = top_players.sort_values(by='red_cards', ascending=False)

# Get the top 5 players
top_5_players = top_players.head(5)

# Display the top 5 players
print(top_5_players)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_players['red_cards'], labels=top_5_players['name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Players by Red Card Counts')
plt.show()
```

```python
INTERPRETATION : 
'''So from the above analysis using a pie chart it is been found that the top 5 players based on red card counts are found and 
Geoff Cameron is leading with 2 red cards with him with an overall percentage of 33.3% and the rest 4 players constituting 16.7% each .'''
```

```python
TOP 5 PLAYERS BY RED CARD COUNTS (V1 : BAR CHART)
```

```python
# Group by player_name and sum the red cards
top_players = file_df.groupby('name')['red_cards'].sum().reset_index()

# Sort the players by total red cards in descending order
top_players = top_players.sort_values(by='red_cards', ascending=False)

# Get the top 5 players
top_5_players = top_players.head(5)

# Display the top 5 players
print(top_5_players)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.barh(top_5_players['name'], top_5_players['red_cards'], color='brown')
plt.title('Top 5 Players by Red Card Counts')
plt.show()
```

```python
TOP 10 PLAYERS BY TOTAL GOALS (V1 : BAR GRAPH)
```

```python
# Group by player_name and sum the goals
top_players = file_df.groupby('name')['goals'].sum().reset_index()

# Sort the players by total goals in descending order
top_players = top_players.sort_values(by='goals', ascending=False)

# Get the top 10 players
top_10_players = top_players.head(10)

# Display the top 10 players
print(top_10_players)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_players['name'], top_10_players['goals'], color='skyblue')
plt.xlabel('Total Goals')
plt.ylabel('Player Name')
plt.title('Top 10 Players by Total Goals')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest scorer at the top
plt.show()
```

```python
INTERPRETATION :
'''Top 10 players based on their goals are represented here using bar graph which is easy and flexible to understand , so based on
the total number of goals hit , the top 10 players are sorted and listed where, Aron Johannsson hits the highest number of goals with a count of
128 goals and tops the list.'''
```

```python
How does key performance metrics( like goals, assists) influence the market value of player
```

```python
# Plotting the data
plt.figure(figsize=(14, 6))

# Scatter plot of Goals vs. Market Value
plt.subplot(1, 2, 1)
sns.scatterplot(x='goals', y='market_value_in_eur', data=file_df)
plt.title('Goals vs. Market Value')
plt.xlabel('Goals')
plt.ylabel('Market Value (in millions)')

# Scatter plot of Assists vs. Market Value
plt.subplot(1, 2, 2)
sns.scatterplot(x='assists', y='market_value_in_eur', data=file_df)
plt.title('Assists vs. Market Value')
plt.xlabel('Assists')
plt.ylabel('Market Value (in millions)')

plt.show()
```

```python
INTERPRETATION :
'''Here goals 0, 1 and 2 have high market value which is evenly spread and comparitively 2, 3, 4 are not spread much and does not have , for the assists
we have the goals 0, 1, 2 where equally it is spread over a range for the market value.'''
```

```python
how does metrics like goals , assists correlate with highest market value fluctuations
```

```python


# Goals vs. Market Value
plt.subplot(1, 2, 1)
sns.regplot(x='goals', y='highest_market_value_in_eur', data=file_df)
plt.title('Goals vs. Market Value')
plt.xlabel('Goals')
plt.ylabel('Market Value (in millions)')

# Assists vs. Market Value
plt.subplot(1, 2, 2)
sns.regplot(x='assists', y='highest_market_value_in_eur', data=file_df)
plt.title('Assists vs. Market Value')
plt.xlabel('Assists')
plt.ylabel('Market Value (in millions)')

plt.show()
```

```python
INTERPRETATION :
'''Here the metrics goals and assists correlation with the highest market value shows that then goals 0 and 1 are highly correlated with highest market 
value where as 4 is least and coming to assists all 3 values 0, 1, 2 are highly correlated with highest market value.'''
```

```python
top 5 players from both the teams comparison in terms of their contributuon to teams overall performance
```

```python
# Calculate Total Contribution
file_df['Total_Contribution'] = file_df['home_club_goals'] + file_df['assists']
# Group by player_name and sum the goals
top_players = file_df.groupby('name')['Total_Contribution'].sum().reset_index()

# Sort the players by total goals in descending order
top_players = top_players.sort_values(by='Total_Contribution', ascending=False)

# Get the top 10 players
top_10_players = top_players.head(10)

# Display the top 10 players
print(top_10_players)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_players['name'], top_10_players['Total_Contribution'], color='purple')
plt.xlabel('Total contribution')
plt.ylabel('Player Name')
plt.title('Top 10 Players by Total Contribution ')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest scorer at the top
plt.show()
```

```python
INTERPRETATION :
''' In the above bar graph it is shown about top 10 players from both the teams based on their overall performance, so from home club it is Fabian Johnson
with a total contribution of 465.'''
```

```python
# Calculate Total Contribution
file_df['Total_Contribution'] = file_df['away_club_goals'] + file_df['assists']
# Group by player_name and sum the goals
top_players = file_df.groupby('name')['Total_Contribution'].sum().reset_index()

# Sort the players by total goals in descending order
top_players = top_players.sort_values(by='Total_Contribution', ascending=False)

# Get the top 10 players
top_10_players = top_players.head(10)

# Display the top 10 players
print(top_10_players)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_players['name'], top_10_players['Total_Contribution'], color='red')
plt.xlabel('Total contribution')
plt.ylabel('Player Name')
plt.title('Top 10 Players by Total Contribution ')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest scorer at the top
plt.show()
```

```python
# Plot distribution of match outcomes based on the number of yellow cards and referee
plt.figure(figsize=(12, 8))
sns.countplot(x='yellow_cards', hue='home_club_position', data=file_df, palette='viridis')
plt.title('Match Outcome Distribution Based on Yellow Cards')
plt.show()

# Plot distribution of match outcomes based on the number of red cards and referee
plt.figure(figsize=(12, 8))
sns.countplot(x='red_cards', hue='home_club_position', data=file_df, palette='magma')
plt.title('Match Outcome Distribution Based on Red Cards')
plt.show()
```

```python
INTERPRETATION :
''' Match outcome based on yellow card and red card is shown above for home club team , it is seen that home club position with the position number 8
is the highest for the number of yellow cards 0, 1 and 2 and for red cards again it is the 8th position but only for the 0 number of red cards.'''
```

```python
# Plot distribution of match outcomes based on the number of yellow cards and referee
plt.figure(figsize=(12, 8))
sns.countplot(x='yellow_cards', hue='away_club_position', data=file_df, palette='viridis')
plt.title('Match Outcome Distribution Based on Yellow Cards')
plt.show()

# Plot distribution of match outcomes based on the number of red cards and referee
plt.figure(figsize=(12, 8))
sns.countplot(x='red_cards', hue='away_club_position', data=file_df, palette='magma')
plt.title('Match Outcome Distribution Based on Red Cards')
plt.show()
```

```python
# Sort the DataFrame by attendance in descending order and select the top 5
df_top5 = file_df.sort_values(by='stadium', ascending=False).head(5)

# Set the size of the figure
plt.figure(figsize=(10, 6))

# Create a barplot for the top 5 attendance counts in different stadiums
sns.barplot(x='stadium', y='attendance', data=df_top5, palette='magma')

# Add title and labels
plt.title('Top 5 Stadiums by Highest Attendance')
plt.xlabel('Stadium')
plt.ylabel('Attendance')

# Show the plot
plt.show()
```

```python
find out the highest attendance count of a football match in different seasons
```

```python
# Find the highest attendance per season
df_max_attendance = file_df.groupby('season').attendance.max().reset_index()

# Set the size of the figure
plt.figure(figsize=(10, 6))

# Create a barplot for the highest attendance in different seasons
sns.barplot(x='season', y='attendance', data=df_max_attendance, palette='plasma')

# Add title and labels
plt.title('Highest Attendance Count in Different Seasons')
plt.xlabel('Season')
plt.ylabel('Highest Attendance')

# Show the plot
plt.show()
```

```python
INTERPRETATION :
''' It is seen that from 2012 - 2019 the attendance was constant throughout consecutively with almost 80,000 counts of people and there is a 
decline with 30,000 count only in the year 2020.
```
