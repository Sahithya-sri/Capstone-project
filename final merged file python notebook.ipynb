```python
import pandas as pd
```

```python
# Load the games data from an Excel file
games_df = pd.read_excel(r'C:\Users\Lenovo\Downloads/games.xlsx')
games_df
```

```python
# Load the games data from an Excel file
game_lineups_df = pd.read_excel(r'C:\Users\Lenovo\Downloads/game_lineups.xlsx')
game_lineups_df
```

```python
# Load the games data from an Excel file
game_events_df = pd.read_excel(r'C:\Users\Lenovo\Downloads/game_events.xlsx')
game_events_df
```

```python
# Load the games data from an Excel file
appearances_df = pd.read_excel(r'C:\Users\Lenovo\Downloads/appearances.xlsx')
appearances_df
```

```python
# Load the games data from an Excel fil
players_df = pd.read_excel(r'C:\Users\Lenovo\Downloads/players.xlsx')
players_df
```

```python
# Merge the 'games_df' DataFrame with the 'game_lineups_df' DataFrame on the 'game_id' column
A_df = pd.merge(games_df, game_lineups_df, on = ['game_id'], how = 'left')

# Merge the resulting DataFrame 'A_df' with the 'game_events_df' DataFrame on the game_id' column
B_df = pd.merge(A_df, game_events_df, on = ['game_id'], how = 'left')

# Merge the resulting DataFrame 'B_df' with the 'appearances_df' DataFrame on the game_id' column
C_df = pd.merge(B_df, appearances_df, on = ['game_id'], how = 'left')

# Merge the resulting DataFrame 'C_df' with the 'players_df' DataFrame on the player_id' column
df1 = pd.merge(C_df, players_df, on = ['player_id'], how = 'left')

df1
```

```python
# to drop the redundant columns
df1 = df1.drop(columns = ['competition_id_y', 'date_y', 'type_x', 'player_id_x', 'player_name_x', 'position_x', 'player_name_y'])
print(df1)
```

```python
# to see the null values present in each column
null_values = df1.isnull().sum()
print(null_values)
```

```python
df1['height_in_cm'] = df1['height_in_cm'].fillna(df1['height_in_cm'].mean())
```

```python
df1['market_value_in_eur'] = df1['market_value_in_eur'].fillna(df1['market_value_in_eur'].mean())
```
